```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xtrain=np.asarray([1,2])
ytrain=np.asarray([200,500])
list1=[]
list2=[]
c1=[]
```

```python
def pred(x,w,b):
    fx=(w*x)+b
    return fx
```

```python
def cost(x,y,w,b):
    total=0
    for i in x:
        total+=(pred(i,w,b)-y)**2
    jwb=1/(2*len(x))*total
    c1.append(jwb)
    return jwb
```

```python
def gredient(x,y,w,b):
    learn_rate=0.2
    sum_der_w=0
    sum_der_b=0
    for i in range(len(x)):
        sum_der_w+=(pred(x[i],w,b)-y)*x[i]
        sum_der_b+=(pred(x[i],w,b)-y)
    temp_w=w-learn_rate*((1/len(x))*sum_der_w)
    temp_b=b-learn_rate*((1/len(x))*sum_der_b)
    w=temp_w
    b=temp_b
    cost(x,y,w,b)
    print(w,b)
    list1.append(w)
    list2.append(b)
    return w,b
```

```python
def linearRegression(x,y):
    m=x.shape
    w=np.zeros(m)
    b=0
    for i in range(1000):
        w,b=gredient(x,y,w,b)
```

```python
linearRegression(xtrain,ytrain)
```

```python
plt.plot(list1)
plt.show()
```

```python
plt.plot(list2)
plt.show()
```
